# Matplotlib 曲线绘制课堂练习

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats

%matplotlib inline
# plt.style.use('Solarize_Light2')
plt.rcParams['font.sans-serif'] = [u'SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams["figure.figsize"] = (8,5)
```

## 1、以 matlab 风格绘制

```python
x_min = 0.0
x_max = 30.0

mean = 9.0 
std = 2.0

# 生成数据的 x-坐标
x1 = np.linspace(x_min, x_max, 100)
# 计算数据的 y-坐标
y1 = scipy.stats.norm.pdf(x1, mean, std)

plt.figure(figsize=(8, 5), facecolor='w', edgecolor='k') 
plt.plot(x1, y1, color='black')
plt.fill_between(x1, y1, color='#89bedc', alpha=1.0);
```

```python
mean = 15.0 
std = 4.0

# 生成数据的 x-坐标
x2 = np.linspace(x_min, x_max, 100)
# 生成数据的 y-坐标
y2 = scipy.stats.norm.pdf(x2,mean,std)

plt.figure(figsize=(8, 5), facecolor='w', edgecolor='k') 
plt.plot(x2, y2, color='black')
plt.fill_between(x2, y2, color='#0b559f', alpha=1.0);
```

## 2、以面向对象方式绘制

```python
fig, ax = plt.subplots(figsize=(8, 5), facecolor='w', edgecolor='k')
ax.plot(x1, y1, color='black')
ax.fill_between(x1, y1, color='#89bedc', alpha=0.8);
```

```python
ax.plot(x2, y2, color='black')
ax.fill_between(x2, y2, color='#0b559f', alpha=0.99)
fig
```

- 或者将代码放一起

```python
import matplotlib.patches as mpatches

label_a = mpatches.Patch(color='#0b559f', label='正态分布 A')
label_b = mpatches.Patch(color='#89bedc', label='正态分布 B')

fig, ax = plt.subplots(figsize=(8, 5))#, facecolor='w', edgecolor='k')
ax.plot(x1, y1, color='black',linewidth=2)
ax.fill_between(x1, y1, color='#89bedc', alpha=0.8)
ax.plot(x2, y2, color='black',linewidth=2)
ax.fill_between(x2, y2, color='#0b559f', alpha=0.9)
plt.legend(handles=[label_a,label_b],fontsize=15);
```

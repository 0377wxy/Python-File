# Matplotlib 折线图绘制

## 一、准备工作

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.rcParams['font.sans-serif'] = [u'SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams["figure.figsize"] = (8,5)
```

## 二、图形与图轴

- 创建一个 figure 图形和一个 axes 图轴；


- plt.style.use 是设置绘图风格，根据个人喜爱设置相应风格。

```python
plt.style.use('fivethirtyeight')

fig = plt.figure()
ax  = plt.axes()
```

- **figure**（类 ``plt.Figure`` 的一个实例）可以视作画布容器，它包含 axes，graphics，text 以及 labels 等。


- **axes**（类 ``plt.Axes`` 的一个实例）是如上看到的画轴（子图），创建好 axes，可以使用 ``ax.plot`` 来绘制数据。

```python
fig = plt.figure()
ax  = plt.axes()

# x 数据
x = np.linspace(0, 10, 100)

# 在一个画轴上绘制两个图形
ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
```

- Matlab 绘制风格

```python
plt.plot(x, np.sin(x));
```

- 一个画轴上绘制多个图形，只要调用多次 plt.plot。

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

## 四、各种参数设置及效果

### 1、调整线条颜色和类型

- 通过 ``plt.plot()`` 设置不同的线条着色。

```python
plt.figure(figsize=(10,5))
plt.plot(x, np.sin(x - 0), color='blue')        # 用名字制定颜色
plt.plot(x, np.sin(x - 1), color='g')           # 颜色简称（rgbcmyk）
plt.plot(x, np.sin(x - 2), color='0.75')        # 0 和 1 之间的灰度值
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # 十六进制分别制定 rgb 三个颜色分量（RRGGBB 每个都是从 00 到 FF）
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB 元组, 取值从 0 到 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # HTML 颜色
```

- 设置线条风格 ``linestyle``。

```python
plt.figure(figsize=(10, 5))

plt.plot(x, x + 0, linestyle = 'solid')
plt.plot(x, x + 1, linestyle = 'dashed')
plt.plot(x, x + 2, linestyle = 'dashdot')
plt.plot(x, x + 3, linestyle = 'dotted')

# 简短方式
plt.plot(x, x + 4, linestyle = '-')  # 实线
plt.plot(x, x + 5, linestyle = '--') # 虚线
plt.plot(x, x + 6, linestyle = '-.') # 虚线加点
plt.plot(x, x + 7, linestyle = ':') # 点
```

- ``plt.plot()`` 也可以同时设置 ``linestyle`` 和 ``color``。

```python
plt.figure(figsize=(10, 5))

plt.plot(x, x + 0, '-g')
plt.plot(x, x + 1, '--c')
plt.plot(x, x + 2, '-.k')
plt.plot(x, x + 3, ':r');
```

### 2、调整坐标轴刻度范围

- 使用 ``plt.xlim()`` 和 ``plt.ylim()`` 调整坐标轴范围。

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

- 可以通过交换 xlim/ylim 中参数大小的顺序来翻转图像。

```python
plt.plot(x, np.sin(x))

plt.xlim(10, 0) # 大值放左边
plt.ylim(1.2, -1.2);
```

- 可以用  ``plt.axis()``  一次性设置  ``[xmin, xmax, ymin, ymax]``（注意这里是 **axis** ，跟 **axes** 要区别开)。

```python
plt.plot(x, np.sin(x))

plt.axis([-1, 11, -15, 15]);
```

- ``plt.axis()`` 方法也可以让包围盒紧贴绘制的图形。

```python
plt.plot(x, np.sin(x))

plt.axis('tight');
```

- 让 $x-$轴和 $y-$轴的单位一致。

```python
plt.plot(x, np.sin(x))
plt.axis('equal');
```

### 3、标题（title）、轴标签（axis label）和图例说明（legend）

```python
plt.plot(x, np.sin(x))

plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("y = sin(x)");
```

- 设置标题、轴标签和图例的字体大小。

```python
import matplotlib

matplotlib.rc('xtick', labelsize=20)
matplotlib.rc('ytick', labelsize=20)
```

- 多个图形同时绘制图例 ``plt.legend()``。

```python
# 定义画布大小
plt.figure(figsize=(10, 5))

# 绘制图形
plt.plot(x, np.sin(x), '-g', label = 'sin(x)')
plt.plot(x, np.cos(x), ':b', label = 'cos(x)')

# 设置标题（title）、轴标签（axis label）和图例说明（legend）
plt.axis('equal')
plt.title('Draw sin(x) and cos(x)', fontsize=18)
plt.legend(loc = "best", fontsize=18);
```

### 4、面向对象绘制方式

- ``plt.plot()`` 改成 ``ax.plot()``


- ``plt.legend()`` 改成 ``ax.legend()``


- ``plt.xlabel()`` 改成 ``ax.set_xlabel()``


- ``plt.ylabel()`` 改成 ``ax.set_ylabel()``


- ``plt.xlim()`` 改成 ``ax.set_xlim()``


- ``plt.ylim()`` 改成 ``ax.set_ylim()``


- ``plt.title()`` 改成 ``ax.set_title()``

- 使用 ``ax.set()`` 一次性设置这些参数。

```python
font = {'family' : 'serif',
        'color'  : 'darkblue',
        'weight' : 'normal',
        'size'   : 18,}
```

```python
# 定义画布大小
plt.figure(figsize=(10, 5))

# 获取当前画轴
ax = plt.axes()

# 在当前画轴上绘制图形
ax.plot(x, np.sin(x), label='sin(x) curve')

# 设置 x 和 y 轴坐标范围和标签
ax.set(xlim=(0, 10), ylim=(-2, 2), xlabel='x', ylabel='sin(x)');

# 设置轴标签和图题及它们的字体
ax.set_xlabel('x = [0, 10]', fontdict=font)
ax.set_ylabel('y = Sin(x)', fontdict=font)
ax.set_title(label='Draw Sin(x)', fontsize=18)

# 设置图例说明
ax.legend(loc = "best", fontsize=18);
```

## 二、基本概念

需要了解的三个主要的类是图形（Figure），图轴（Axes）以及坐标轴（Axis）。

- 图形（Figure）

它指的就是你看到的整个图形窗口。同一图形中可能有多个子图（图轴）。在上面的示例中，在一个图形中有四个子图（图轴）。

- 图轴（Axes）

图轴就是指图形中实际绘制的图。一个图形可以有多个图轴，但是给定的图轴只是整个图形的一部分。在上面的示例中，我们在一个图形中有四个图轴。

- 坐标轴（Axis）

坐标轴是指特定图轴中的实际的 x-轴和 y-轴。

## 四、两套绘制机制

- MATLAB 风格的界面


- 面向对象方式的界面。

### 1、MATLAB 风格

MATLAB 风格采用 pyplot (``plt``)

```python
# 创建一个 plot 画布
plt.figure()

# 创建一个绘图子窗口
plt.subplot(2, 1, 1) # (行数, 列数, 第几个绘制面板)
plt.plot(x, np.sin(x))

# 创建第二个绘图子窗口以及绘制相应的图
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x), '-r');
```

- 这种方式可以使用 ``plt.gcf()`` 获取当前图形 (get current figure)


- 以及当前画轴 ``plt.gca()`` (get current axes)。


- 缺点就是不能返回到前面图形上继续绘制。

### 2、面向对象的界面

通过显式调用图形 ``Figure`` 和图轴 ``Axes`` 对象的方法来绘制不同的图。

1、创建一个带若干个 subplots 的 figure

```python
# ax 表示 Axes 对象
fig, ax = plt.subplots(2)

# 调用相应对象的方法 plot()
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

这种方法允许到前面图形上继续绘制新的图。

```python
ax[0].plot(x, np.cos(x), 'r')
fig
```

可以看到两种方式绘制形式上的主要不同是分别调用 ``plt.plot()`` 和 ``ax.plot()``。

两种方式我们都要掌握。

2、先创建图形 figure，再添加图轴 axes

```python
plt.style.use('bmh')
```

```python
x = np.linspace(0, 5, 10)
y = x ** 2
```

```python
# 创建图形，并指定窗口大小
fig = plt.figure(figsize=(12,6))

# 左, 底, 宽度, 高度 (范围是 0 到 1)
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 
# inset axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) 

# 大图
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Figure')

# 内嵌小图
axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('Inset');
```

```python
axes2.plot(x, y, 'b')
fig
```

### 3、绘制 3D 图形的例子

```python
plt.style.use('classic')
```

```python
import random
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from mpl_toolkits.mplot3d import Axes3D

mpl.rcParams['font.size'] = 10

# 设定 figure 的大小
fig = plt.figure(figsize=(12,8))

# 添加图轴
ax = fig.add_subplot(111, projection='3d')

for z in [2015, 2016, 2017, 2018, 2019]:
    xs = range(1,13)
    ys = 1000 * np.random.rand(12)

    color = plt.cm.Set2(random.choice(range(plt.cm.Set2.N)))
    ax.bar(xs, ys, zs=z, zdir='x', color=color, alpha=0.8)

ax.xaxis.set_major_locator(mpl.ticker.FixedLocator(xs))
ax.yaxis.set_major_locator(mpl.ticker.FixedLocator(ys))

ax.set_xlabel('Year')
ax.set_ylabel('Month')
ax.set_zlabel('Sales Net [usd]')
```

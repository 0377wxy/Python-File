```python
def Determinant_3(arr):
    res=arr[0,0]*(arr[1,1]*arr[2,2]-arr[1,2]*arr[2,1])- \
        arr[0,1]*(arr[1,0]*arr[2,2]-arr[1,2]*arr[2,0])+ \
        arr[0,2]*(arr[1,0]*arr[2,1]-arr[1,1]*arr[2,0])
    return res

def Determinant_5(arr):
    res=np.ones((3,3))
    for i in range(3):
        for j in range(3):
            res[i,j]=Determinant_3(arr[i:i+3,j:j+3])
    return res

def Determinant_all(arr):
    res =np.ones((5,3,3))
    for i in range(5):
        res[i]=Determinant_5(A[i])
    return res

```
